```python
import astropy.io.fits as fits
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
fold = '/home/ryan/projects/CHARMKASTR/raw/data-2017-08-12-shane-Adam.Burgasser/red/test1035/science/'
```

```python
hdu = fits.open(fold+'r1033.fits')
data = hdu[0].data
hdu.close()
```

```python
plt.imshow(data)
```

```python
xpk = []
impc = []
for i in range(len(data[:,0])):
    slc = data[i,:]
    xpk.append(numpy.argmax(slc))
    if xpk[-1] > 150 and xpk[-1] < 165:
        impc.append(slc[xpk[-1]-10:xpk[-1]+10])
plt.plot(numpy.sum(impc,axis=1))
```
